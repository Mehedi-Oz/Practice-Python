```python
# Order-Agnostic Binary Search


def binary_search(start, end, nums, target):

    if nums[start] == nums[end]:
        return 0 if nums[0] == target else -1

    is_ascending = nums[start] < nums[end]

    while start <= end:
        mid = (start + end) // 2

        if nums[mid] == target:
            return mid

        if is_ascending:
            if nums[mid] < target:
                start = mid + 1
            else:
                end = mid - 1
        else:
            if nums[mid] > target:
                start = mid + 1
            else:
                end = mid - 1
    return -1


def order_agnostic_search(nums, target):
    start, end = 0, len(nums) - 1

    return binary_search(start, end, nums, target)


nums = [90, 79, 72, 60, 55, 35, 20, 15]
# nums = [1, 4, 5, 6, 9]

target = 1
print(f"Order-Agnostic : {order_agnostic_search(nums, target)}")
```

```python
# Ceiling of A Number


def floor_of_target(nums, target):
    first, last = 0, len(nums) - 1

    if target<nums[first]:
        return -1

    while first <= last:
        mid = (first + last) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            first = mid + 1
        else:
            last = mid - 1

    return nums[last]


def ceiling_of_target(nums, target):
    first, last = 0, len(nums) - 1

    if target>nums[last]:
        return -1

    while first <= last:
        mid = (first + last) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            first = mid + 1
        else:
            last = mid - 1

    return nums[first]


nums = [1, 2, 8, 10, 10, 12, 19]
target = 0
print(f"Ceiling Of the target : {ceiling_of_target(nums, target)}")
print(f"Floor Of the target : {floor_of_target(nums, target)}")
```

```python
# 744. Find Smallest Letter Greater Than Target
# https://leetcode.com/problems/find-smallest-letter-greater-than-target/description/


def finding_smallest_letter(letters, target):
    first, last = 0, len(letters) - 1

    while first <= last:
        mid = (first + last) // 2

        if letters[mid] <= target:
            first = mid + 1
        else:
            last = mid - 1

    return letters[first % len(letters)]


letters = ["c","f","j"]
target = "b"
print(f"Output: {finding_smallest_letter(letters, target)}")
```

```python
# Search in Rotated Sorted Array with duplicates


def binary_search(first, last, arr, target):
    while first <= last:
        mid = (first + last) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            first = mid + 1
        else:
            last = mid - 1

    return -1


def pivot_in_rotated_array(nums):
    first, last = 0, len(nums) - 1

    while first <= last:
        mid = (first + last) // 2

        if mid < last and nums[mid] > nums[mid + 1]:
            return mid
        if mid > first and nums[mid] < nums[mid - 1]:
            return mid - 1
        if nums[first] == nums[mid] == nums[last]:  # Handle duplicate elements
            first += 1
            last -= 1
        elif nums[first] <= nums[mid]:  # Left part is sorted, pivot is on right
            first = mid + 1
        else:  # Right part is sorted, pivot is on left
            last = mid - 1
    return -1


def target_search(nums, target):
    pivot = pivot_in_rotated_array(nums)

    if pivot == -1:
        return binary_search(0, len(nums) - 1, nums, target)
    elif nums[pivot] == target:
        return pivot
    elif target >= nums[0]:
        return binary_search(0, pivot - 1, nums, target)
    elif target < nums[0]:
        return binary_search(pivot + 1, len(nums) - 1, nums, target)


nums = [1,1,1,1,1,2]
target = 2
print(f"Index: {target_search(nums, target)}")
```

```python
# 410. Split Array Largest Sum
# https://leetcode.com/problems/split-array-largest-sum/description/


def largest_sum(nums, k):
    start = max(nums)
    end = sum(nums)

    while start < end:
        mid = (start + end) // 2

        current_sum = 0
        pieces = 1

        for num in nums:
            if current_sum  + num > mid:
                current_sum  = num
                pieces += 1
            else:
                current_sum  += num

        if pieces > k:
            start = mid + 1
        else:
            end = mid

    return end


nums = [7, 2, 5, 10, 8]
k = 2
print(largest_sum(nums, k))
```
